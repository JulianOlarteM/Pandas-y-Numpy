```python
import pandas as pd
import numpy as np
```

```python
DF1=pd.DataFrame({'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3']})
```

```python
DF2=pd.DataFrame({'A': ['A4','A5','A6','A7'],
 'B': ['B4','B5','B6','B7'],
 'C': ['C4','C5','C6','C7'],
 'D': ['D4','A5','A6','A7']})
```

PARA CONCATENAR ESTO SE HACE:

```python
pd.concat([DF1,DF2])
```

PARA COREGIR LOS INDICES TENEMOS

```python
pd.concat([DF1,DF2],ignore_index=True)
```

```python
pd.concat([DF1,DF2],axis=1)
```

MERGE


```python
izq = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })
```

```python
der = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })
```

```python
izq.merge(der, on = 'key')
```

```python
izq = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })

der = pd.DataFrame({'key_2' : ['k0','k1','k2','k3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })

```

```python
izq.merge(der,left_on='key',right_on='key_2')
```

Modifiquemos de nuevo las lista para ver que pasa si uno de los indices no coincide. 

```python
izq = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3'],
 })

der = pd.DataFrame({'key_2' : ['k0','k1','k2',np.nan],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3'],
 })
```

```python
der
```

como modificamos las llaves del segundo dataframe, al hacer el marge solo me trae las llaves que coinciden en ambos data frames.

```python
izq.merge(der,left_on='key',right_on='key_2')
```

Como hacemos para traer todo lo de la derecha o todo lo de la izquierda sin importan que coincida con el otro dataframe?

```python
izq.merge(der,left_on='key',right_on='key_2',how='left')
```

```python
izq.merge(der,left_on='key',right_on='key_2',how='right')
```

JOIN - Index match

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
                    'B': ['B0','B1','B2',]},index=['k0','k1','k2'])

der = pd.DataFrame({'C': ['C0','C1','C2'],
                    'D': ['D0','D1','D2',]},index=['k0','k2','k3'])
```

```python
izq.join(der)
```

```python
izq.join(der,how='inner')
```

```python
izq.join(der,how='left')
```

```python
izq.join(der,how='right')
```

```python
izq.join(der,how='outer ')
```
