Filtrando datos con loc y iloc

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv',sep=',',header=0)
```

```python
df_books[0:4]
```

```python
df_books[['Name', 'Author','Year']]
```

Pero ahora,  ¿como hago para filtrar datos de una combinacion entre filas y columnas?. Para esto se utiliza *Loc*

```python
df_books.loc[0:4]
```

```python
df_books.loc[0:4, ['Name','Author']]
```

```python
df_books.loc[:, ['Reviews']] *-1
```

```python
df_books.loc[:, ['Author']] == 'JJ Smith'
```

Funcion Iloc

```python
df_books.iloc[:]
```

```python
df_books.iloc[:,0:3]
```

```python
df_books.iloc[1,3]*-1
```

```python
df_books.iloc[:3,2:]
```
