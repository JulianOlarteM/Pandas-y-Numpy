```python
import numpy as np
```

La manera tradicional de creacion de listas puede ser como:

```python
list(range(0,10))
```

pero numpy tambien nos permite la creadcion de arrays desde cero, aprovechando todos los beneficos del modulo.

```python
np.arange(0,10)
```

```python
np.arange(0,20,2)
```

```python
np.zeros(3)
```

```python
np.zeros((10,5))
```

```python
np.ones((10,5))
```

```python
np.linspace(0,10,40)
```

Para los amantes del algebra, tambien es posible:

```python
np.eye(4)
```

Para generar valores aleatorios:

```python
np.random.rand()
```

```python
np.random.rand(4)
```

```python
np.random.rand(4,4)
```

Para generar un numero aletario entre un rango:

```python
np.random.randint(1,15)
```

```python
np.random.randint(1,100, (10,10))
```
