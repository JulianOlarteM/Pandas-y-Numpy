Veamos como creo otro arreglo diferente al original, y al modificarlo se modifica tambien el original.

```python
import numpy as np
```

```python
arr = np.arange(0,11)
arr
```

```python
trozo_arr = arr[:6]
trozo_arr
```

```python
trozo_arr[:] = 0
trozo_arr
```

```python
arr
```

Para solucionar esto, exite el comando copy() 

```python
arr_copy = arr.copy()
```

```python
arr_copy
```

```python
arr_copy[:] = 100
arr_copy
```

```python
arr
```
