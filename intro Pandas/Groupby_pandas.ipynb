```python
import pandas as pd
```

```python
df_books=pd.read_csv('/content/bestsellers-with-categories.csv',sep=',',header=0)
df_books.head(2)
```

Aca agrupo la categoria Author, y le pregunto cuantos registro ahi por cada autor

```python
df_books.groupby('Author').count()
```

Para agrupar la categoria author y que me muestre la media de cada caracteristica de la matriz

```python
df_books.groupby('Author').mean().head(2)
```

Tambien se puede agrupar por author  mostar los valores minimos por categoria

```python
df_books.groupby('Author').min().head(2)
```

```python
df_books.groupby('Author').max().head(2)
```

ahora realizamos una suma de los valores numericos por categoria,

```python
df_books.groupby('Author').sum()
```

Podemos tambien obtener la suma de valores, pero ahora como agrupamos la categoria "Author", esta se vuelve nuestro indice, y podemos buscar el indice que necesitemos.

```python
df_books.groupby('Author').sum().loc['William Davis']
```

Si queremos agrupar por ejemplo la categoria "Author", peor no queremos que esta sea el indice, podemos realizar lo siguiente. 

```python
 df_books.groupby('Author').sum().reset_index()
```

Agrupamiento por funciones de agregacion : 
En este caso agrupamos la categoria autores, y a ese agrupamiento le agregamos una funcion como puede ser min(), max(),mean(),sum(), para cada una de las demas categorias. 

```python
df_books.groupby('Author').agg(['min','max'])
```

Ahora tambien podemos generar diccionarios donde las llaves son categorias llamadas del data set, y a estas aplicarles funciones especicfas. En este caso se llamo por autor a el minimo y el maximo de revews por libro, y a la sumatoria de los volres de 'User Rating' por cada autor, 

```python
df_books.groupby('Author').agg({'Reviews': ['min','max'], 'User Rating' : ['sum']})
```

Otra cosa importante de entender es que podemos decidir que datos mostrar los  hacer operaciones a estas categorias.

```python
df_books.groupby(['Author','Year']).count()
```

Un manejo mas estructurado podria ser:

```python
df_books.groupby('Author').agg({'Year':lambda x: [i-2000 for i in x], 'Reviews':lambda x: sum([i**2 for i in x])})
```
